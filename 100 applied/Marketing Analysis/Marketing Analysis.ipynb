### Peak Keyword during Covid VS Current

```python
three_keywords['month_date'] = pd.to_datetime(three_keywords['month'], format='%Y-%m')
three_keywords.head()
```

```python

plt.plot(three_keywords['month_date'],three_keywords['home_workout_worldwide'], color='r', label='home_workout_worldwide')
plt.plot(three_keywords['month_date'],three_keywords['gym_workout_worldwide'], color='g', label='gym_workout_worldwide')
plt.plot(three_keywords['month_date'],three_keywords['home_gym_worldwide'], color='b', label='home_gym_worldwide')
plt.legend()
plt.title('Popularity of the keyword')
plt.show()
```
