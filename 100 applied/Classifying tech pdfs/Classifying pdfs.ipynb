```python
from pathlib import Path
from isbnlib import get_isbnlike, meta,is_isbn10, is_isbn13, get_canonical_isbn, desc

import requests
from langchain_community.document_loaders import PyPDFLoader

path = '/Users/el_fer/Documents/Libros/Libros Tecnicos'
folder_path = Path(path)

pdf_files = list(folder_path.rglob("*.pdf"))

test_pdf = pdf_files[0]
```

```python
from typing import List
from collections import Counter

def get_valid_isbns(isbns: List[str]) -> List[str]:
    return [get_canonical_isbn(isbn, output='bouth') for isbn in isbns if is_isbn10(isbn) or is_isbn13(isbn)]
```

```python
import json 
infos={}
isbn= None 

for pdf in pdf_files:
    print(f'PROCESSING {pdf}')
    infos[pdf] = {}
    loader = PyPDFLoader(pdf)
    pages = []
    try: 
        async for page in loader.alazy_load():
            pages.append(page)
        
        for page in pages[:10]: 
            isbns = get_isbnlike(page.page_content, level='normal')
        
            if len(isbns) == 1:
                isbn = isbns[0]
            elif len(isbns)>1: 
                isbn = Counter(get_valid_isbns(isbns)).most_common(1)[0][0]
            else:
                # no isbn on this page
                pass
    
            if isbn:
                print(f'Got the isbn {isbn}')
                infos[pdf]['isbns'] = isbns
                try:
                    infos[pdf]['description'] = desc(isbn)
                    infos[pdf]['metadata'] = meta(isbn)
                except Exception as e:
                    print(f'Error processing {pdf}: ', e)
                    infos[pdf]['error'] = e
                    
                isbn = None
                break
    
        print(infos[pdf])
        print('-------')
    except AttributeError as ae:
        print(f'Impossible to read {pdf}', ae)

print(infos)
```
