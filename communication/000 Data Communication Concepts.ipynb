## Fundamentals of Storytelling

The data storytelling road:

  Story > Tech or not tech > data > viz > present

Why do we need stories?

The best results could have no impact without proper presentation

The main challenge is to convince stakeholders that things have to be done differently, specially if these stakeholders are non technical or dont feel confortable with math or statistics.

## What is data storytelling?

Data storytelling is the practice of building a narrative around a set of data and its accompanying visualizations to help convey the meaning of that data in a powerful and compelling fashion

Anecdotes -> peoples impagination

Stories are more memorable than metrics

And they add value giving context to the data

The attention of the audience is then captured facilitating the decision making and thereforem drive change through actions.

## How to build an effective data story?abs
There are two main concepts that will help us articulate our story in a clear and concice manner:

#### The 3 minutes story
What would we focus on if we have only 3 minutes to tell our story?

#### The Big Idea
Unique point of view and one single sentence

These elements will help us creating a story thats:
- Insightful
- Explanatory
- Concise

### Data
To build an effective story we should include only findings from our models and analysis that apply to the situation. They have to be relevant. Our story has to be accurate and reliable, based on our results. our story has to have actionable insights.

### Narrative
A good story has to be compelling and easy to understand including only the key points required to drive change.

Avoid including disconnected facts and try to keep it turning around a central insight, taking into account our intended audience and their background so we can clarify facts for them. 

Every data story follows a linear sequence, so each data point is connected till the conclusion is reached.

### Visuals
Graphs should be simple, engaging and not misleading

## How technical our audience is?

For the sake of simplification we are considering here technical or non technical audiences, while the reality doesnt go black or white but more like a continuum.

Data professionals like to care about the methods, while the audience tends to care more about results and implications.

### Translating technical results into stories

In order to explain the technical results to a non technical audience in an engaging and easy to understand fashion we need to use common terms that will enable decision making and drive change.

There are several strategies to accomplish this goal:

- What do they need to know to understand? -> Be conversational
- What level of information do they need? -> Our story should be able to make the audience understand what our project is about

## Compelling Narrative

After finding insights worth sharing, we have to build a meaningful narrative for the target audience. Prioritize key points so we can drive change.

We need a compelling narratve or a description of connected events that organizes information to engage the audience and make them care for the results of the information shared.

## Narrative structure

### 1. Background
- What motivated the analisis
- What changed?
- Who is the focus of the analysis?
  - Customers? Employees? Anyone else?
 
### 2. Insight 1
- We can provide them elements that contributed to the problem
- But only relevant information

### 3. Insight 2
- Add supporting evidence
- Help better explain the cause of the problem

### 4. Climax
All the evidences should lead to the climax. The central finding of our analysis: it should state clearly what could happen if nothing changes.

### 5. Next steps
Potential solutions to the identified problem
Course of action from a proactive point of view.

## Building the narrative 
We can focus on the change over time or in some interesting correlation or similarities. Clustering, creating groups could be interesting too.

# Selecting the right data

Selecting the right data implies including enough contextual insights in a story to better support the main point without overloading with information: we include the minimal information supporting our story.

## Knowing our stakeholders
We should adapt to the different kinds of stakeholders interested in our project. A stakeholder can be any person or group of persons interested in the project outcome or a decision or activity derived from it. 
They can be tech or not tech.

It can be really helpful to identify the different personas by describing their interests and knowledge.

## Audience Skepticism
There are different levels of skepticism.

We should be ready to add more data if our reasoning is bringing up doubts. A common roadmap for this is:

1. convince yourelf
2. convince a friend
3. convince a skeptic




## Showing variations of data

Differences can be expressed with:
- absolute values (for 1 variable is ok)
- growth rate (recommended for the case of 2+ variables)

Pay attention to the global quantity, since growth rates tend to look bigger on small base figures.

### Ratios
Ratios can be also used. Ratios help normalize values, what leads to better comparisons.

### Aggregates
Aggregates like totals, counts, means, median, sums are helpful too.

### P-value
Statistical proof that our results are statistically significant is often required A p-value below 0.05 is considered a proof of statistical significance.

The p-value can not be considered a proof of evidence: we can reject the null hypothesis, but we cannot conclude that it is false.

Yet, knowing how confusing the p-value can be, its preferable to introduce other kind of metrics easier to understand.


## Visualizations for different audiences

2 Things to consider when picking a visualization technique are the expertise and the familiarity of the audience with the visuals.

Best practices:
- Pareto principle: the less relevant data can be aggregated
- Approachable and engaging visuals
- How many insights and how quickly they can get
- Less is more

### The McCandless method (https://akshaysbabu.medium.com/what-i-understood-from-reading-david-mccandless-the-art-science-of-data-presentation-7982d4d74cef)
1. Introduce visualization by name (graph short, clear and obvious headline). Axis lables could be used too
2. Anticipate audiences questions
3. State insights
4. Help the audience relate: importance, action items...

- Introduction
  - Purpose
  - Contextual information, motivation
  - Questions of analysis
- Body
  - Data used
  - Methods
  - Analysis
    - Visuals
  - Results
    - Description and visuals
- Conclusions
  - Restate question
  - Summarize important results
  - Add recommendations and next steps

In a business context, we have to adapt the previous report. A good rule is:
- 1 page of abstract
- 3 pages max of executive summary
- 25 pages max of detail

Keep in mind the audience


## Reproducibility and References
An important point is that the results of the project have to be reproducible independently of the technology used.
This will allow peer review, reutilisation and focus on new challenges.

Best practices:
- Keep track of how the results were produced.
  - Well documented scripts
  - Version control
- Avoid manual data manipulation
  - Data versioning
  - Store raw data and intermediate steps
  - Adapt and resolve problems
- Control randomness
  - Set random seed
- Interpretability
  - Understand the cause of a decision or predict model results
  - Stor with compelling narrative
  - Link with reproducibility
- Cite bibliography correctly (In a business context this part is way more relax)

## Planning an oral presentation

Before start creating slides, lets plan!
Its important to define the presentation structure before anything else:
- Purpose of the presentation:
  - Is it an informative presentation or an instructional one? Are we trying to pursuade stakeholders?
- Who the audience is?
  - Is it technical or executives?
  - How big it is?
  - Is it a workshop?
- Message to convey:
  - What is the central message? What do we want to stick?
  - The opening statement should capture audiences attention and already mention the central message
  - Keep the central message one sentence.
  - The closing statement should sum up the presentation and strengthen the central message

Once the previous points are clear we can start defining the structure of the oral presentation:
- Introduction
  - Provide background information
  - Catch audience attention
  - Glimpse of presentation content
- Methods, analysis and model outputs (this whole section would be deeply determined by who our audience is)
- Conclusions and takeaways
  - Refers back to introduction
  - Contains call to action statement and or next steps

Now its time to outline graphs and visuals and the sections (5 or less smaller parts)

Keep time in mind!!!

## Building presentation slides

Our slides should support the story we want to convey. Its better more slides with small amount of information rather than overwhelming content. They should be refined. Instead of counting the time on each slide or the number of slides its better to keep one message per slide.

The usage of colors can be useful but distracting as well. Dont use more than the required. Keep a good contrast between the text and the background.

Are there any visually impaired persons in the audience?

Fonts are important too.

Short and big texts rather than small and long.

Write the main ideas, not the whole explanation. Its a support, not the main thing.

Headlines!

Layering slides is a good approach to split long slides into several parts.

Replace long sentences with graphs.

Highlight graphs or parts of it.

PRACTICE! PRACTICE! PRACTICE!

Anticipate follow-up questions

Rehearsal: stand up, use the slides, speak out loud, detect distracting patternsm find linking statements

Be aware of emotions.
- Confidence vs unsure

Short attention span of the audience. Talk to the audience to keep them engaged. Ask stuff. Develop a relationship.

Dont use more than the allocated time.

Watch out the pace.

Open up for questions.


## Avoid common mistakes

- Presentations going too long
- Not reserving time for questions
- Forgetting to mention the purpose
- Dont wait to list the findings
- Monologue
- Strong introduction
- Not engaging with the audience
- Not reiterating the main idea
- Not paying attention to the body language
